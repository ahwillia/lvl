```python
import numpy as np
import matplotlib.pyplot as plt
from lvl.glm import poiss_reg
```

```python
%%time
rs = np.random.RandomState(1234)

n_obs = 100000
n_features = 100
n_dep_vars = 100

X = 0.1 * rs.randn(n_obs, n_features)
W_true = 0.1 * rs.randn(n_dep_vars, n_features)
Y = rs.poisson(np.exp(X @ W_true.T)).astype(float)

_wx = X @ W_true.T
true_loss = (np.exp(_wx).sum() - (_wx * Y).sum()) / n_obs

plt.hist(Y.ravel(), 10);
```

```python
%%time
W_sgd, sgd_hist = poiss_reg(X, Y, method="sgd", patience=10000, max_epochs=1, num_ss=4, W=W_sgd.copy())
```

```python
%%time
W_sgd_avg, sgd_avg_hist = poiss_reg(X, Y, method="sgd_avg", patience=10000, max_epochs=10)
```

```python
plt.plot(sgd_hist, label="sgd")
plt.plot(sgd_avg_hist, label="sgd_avg")
plt.axhline(true_loss, dashes=[2, 2], color='r')
```
